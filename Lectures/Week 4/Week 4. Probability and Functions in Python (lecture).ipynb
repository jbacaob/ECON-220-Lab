## Step 3. Defining a more involved function

```python
def simulate_rolls(num_tosses, target_number):
    count = 0
    for _ in range(num_tosses):
        roll = roll_dice()
        print(f"The dice fell on: {roll}")
        if roll == target_number:
            count += 1
    return count / num_tosses
```

Simulate rolling a dice multiple times and calculate empirical probability of getting a specific number

```python
num_tosses = 10000
target_number = 3
empirical_prob = simulate_rolls(num_tosses, target_number)
```

## Step 4. Law of large numbers

```python
def convergence_rolls(num_tosses, target_number):
    empirical_probs = [] # Initialize empty list to store empirical probabilities
    count = 0

    for i in range(1, num_tosses + 1):
        roll = roll_dice()
        if roll == target_number:
            count += 1
        empirical_prob = count / i
        empirical_probs.append(empirical_prob) # Append current empirical probability to the list
    
    return empirical_probs
```

```python
num_tosses = 10000
target_number = 3
empirical_prob_list = convergence_rolls(num_tosses, target_number)
empirical_prob_list
```

## Visualize the empirical probabilities

```python
plt.figure()
plt.plot(range(1, num_tosses + 1), empirical_prob_list, label='Empirical Probability', color='blue')
plt.axhline(y=1/6, color='red', linestyle='--', label='Theoretical Probability (1/6)')
plt.xlabel('Number of Tosses')
plt.ylabel('Empirical Probability of Rolling a 3')
plt.title('Is the empirical probability getting closer to the theoretical probability?')
```
