## Step 2. Create a function

Now, we weant to create a function that tells us the result of some random die toss. A die toss can result in one of the following numbers: 1, 2, 3, 4, 5, or 6. 

```python
def roll_dice():
    return random.randint(1, 6)

# Implement the function once
roll_dice()
```

What is the _theoretical probability_ of getting a 3 in a die toss? If the die is fair, the theoretical probability of getting a 3 is 1/6.

```python
theoretical_prob = 1/6
```

Let's test this result by simulating multiple die rolls. To do this, we start with the first trial (count = 0 since Python starts counting from 0). Then, we specify the number of trials we want to run (e.g., 1000) and the target number we are looking at. Our target number is 3.

```python
def simulate_rolls(num_tosses, target_number):
    count = 0
    for _ in range(num_tosses):
        roll = roll_dice()
        print(f"The dice fell on: {roll}")
        if roll == target_number:
            count += 1
    return count / num_tosses

# Simulate 1000 dice rolls and calculate the empirical probability of getting a 3
num_tosses = 1000
target_number = 3
empirical_prob = simulate_rolls(num_tosses, target_number)
```

What is the empirical probability that we obtained in the experiment?

```python
empirical_prob
```

Our empirical probability is relatively close to the theoretical probability of 1/6 (0.1667).

A natural question is: what happens if we increase the number of trials? Do we get closer to the theoretical probability? Let's try it out!

```python
def convergence_rolls(num_tosses, target_number):
    empirical_probs = [] # Initialize an empty list to store empirical probabilities
    count = 0

    for i in range(1, num_tosses + 1):
        roll = roll_dice()
        if roll == target_number:
            count += 1
        empirical_probs.append(count / i) # Append empirical probability at this trial
    
    return empirical_probs

# Simulate 1000 dice rolls and track empirical probabilities
num_tosses = 1000
target_number = 3
empirical_probs = convergence_rolls(num_tosses, target_number)

empirical_probs
```

Rule of thumb: a plot is worth a thousand words. Let's visualize the empirical probability as we increase the number of trials.

```python
plt.figure()
plt.plot(range(1, num_tosses + 1), empirical_probs, label='Empirical Probability', color='blue')
plt.axhline(y=theoretical_prob, color='red', linestyle='--', label='Theoretical Probability (1/6)')
plt.xlabel('Number of Rolls')
plt.ylabel('Probability of Rolling a 3')
plt.title('Is the empirical probability getting closer to the theoretical probability?')
plt.legend()
plt.show()
```

The answer is yes! As we increase the number of trials, the empirical probability gets closer and closer to the theoretical probability of 1/6. This is what we call the **Law of Large Numbers**.
