### Loading packages

After installing the required packages, the first step is to load them. In our class, three packages will be essential: Pandas, NumPy, and Matplotlib. There are tons of other packages that you can use, but these three are the most common and useful.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

The command ```import``` is used to load packages in Python. You can also use the command ```as``` to give a package a shorter name. For example, we will use ```pd``` as a short name for Pandas, ```np``` for NumPy, and ```plt``` for Matplotlib.

Let's now import the dataset. For this, we first specify the path to the data file. In this case, we will use a CSV file.

````python
# Specify the path to the data file
path = "https://raw.githubusercontent.com/jbacaob/ECON-220-Lab/refs/heads/main/Lectures/Week%202/starwars.csv"

# Load the dataset
data = pd.read_csv(path)
````

The key ```#``` is used to write comments in Python. Comments are ignored by the interpreter and are useful to explain what the code does. Use it whenever you can, especially in your assignments.

### Exploring the dataset

We can use the command ```head()``` to see the first rows of the dataset. This is useful to get a quick overview of the data.

````python
data.head(10)  # Display the first 10 rows of the dataset
````

If we instead want to check the last rows of the dataset, we can use the command ```tail()```.

````python
data.tail(10)  # Display the last 10 rows of the dataset
````

Let's continue our exploration by checking the dimensions of the dataset, the column names, and the data types of each column.

````python
data.shape  # Get the dimensions of the dataset (rows, columns)
````

````python
data.columns  # Get the column names of the dataset
````

````python
data.dtypes  # Get the data types of each column
````

Now we want to know the average height of the characters in the dataset. We can use the command ```mean()``` to calculate the mean of a column. If we want more details about this variable, we can use the command ```describe()```, which provides a summary of the statistics for the column.

**Important**. According to the scales of measurement, height is a ratio variable since it has a true zero point (0 cm means no height) and the ratios between values are meaningful (e.g., 180 cm is twice as tall as 90 cm).

````python
data['height'].mean()  # Calculate the mean height of characters
````

````python
data['height'].describe()  # Get summary statistics for the height column
````

For categorical variables, we can use the command ```value_counts()``` to see the frequency of each category. For example, we can see how many eye colors have the characters in the dataset.

````python
eyecolor_count = data['eye_color'].value_counts()  # Get the frequency of each species in the dataset
eyecolor_count  # Display the frequency counts
````

### Visualizing the data

One image is worth a thousand words. Let's visualize the frequency of each species in the dataset using a bar chart.

````python
plt.figure(figsize=(7, 4))
plt.bar(eyecolor_count.index, eyecolor_count.values)
plt.title('Frequency of Eye Colors')
plt.xlabel('Eye color')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
````

Alternatively, we could have used a pie chart to visualize the frequency of each species in the dataset.

````python
plt.figure(figsize=(14, 6))
plt.pie(eyecolor_count.values, labels=eyecolor_count.index, autopct='%1.1f%%', startangle=140)
plt.title('Frequency of Eye Colors')
plt.xlabel('Eye color')
plt.ylabel('Count')
plt.show()
````

### Handling data (basics)

The last pie chart does not look very good because there are too many categories. A bar chart is more appropriate in this case. Suppose that we are only interested in eye colors with more than 5 characters. How can we filter the data to only include these categories? We use indexing.

````python
freq_eye_colors = eyecolor_count[eyecolor_count > 3]  # Filter eye colors with more than 5 characters
freq_eye_colors  # Display the filtered frequency counts
````

````python
plt.figure(figsize=(14, 6))
plt.pie(freq_eye_colors.values, labels=freq_eye_colors.index, autopct='%1.1f%%', startangle=140)
plt.title('Frequency of Eye Colors')
plt.xlabel('Eye color')
plt.ylabel('Count')
plt.show()
````

This chart looks way better. We can do more advanced subsetting. For example, we want to find the different homeworlds of human and droid characters.

````python
condition = (data['species'] == 'Human') | (data['species'] == 'Droid')
columns_to_keep = ['species', 'homeworld']
filtered_data = data[columns_to_keep][condition]  # Filter the dataset for human characters
filtered_data
````

We can sort the filtered data by species to appreciate the differences.

````python
filtered_data.sort_values(by='species')  # Sort the filtered data by species
````
